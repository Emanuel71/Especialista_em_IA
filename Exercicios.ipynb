Você está desenvolvendo uma funcionalidade para um aplicativo de lanchonete. O sistema precisa calcular o valor total de um pedido com base na quantidade de itens solicitados.

O cliente pode pedir:

Hambúrguer: R$ 12,00<br>
Batata frita: R$ 7,00<br>
Refrigerante: R$ 5,00<br><br>
Sua tarefa é criar um algoritmo em linguagem natural ou fluxograma que armazene os dados necessários em variáveis e faça o cálculo do valor total com base nas quantidades informadas.

```python
cardapio = {'Hamburguer':12.00, 
            'Batata_frita':7.00,
            'Refrigerante':5.00}

def calcular_total(pedido):
    total = 0
    for item in pedido:
        if item in cardapio:
            total += cardapio[item]
        else:
            print(f"Item '{item}' não encontrado no cardápio.")
    return total
```

```python
calcular_total(['Hamburguer', 'Batata_frita', 'Refrigerante', "Hamburguer"])
```

Você está desenvolvendo um sistema para uma empresa de delivery. O valor da taxa de entrega depende da distância até o cliente e se o pedido foi feito em um dia de chuva.

As regras são:

Para entregas até 5 km, a taxa é R$ 5,00.<br>
Entre 5 e 10 km, a taxa é R$ 8,00.<br>
Acima de 10 km, a taxa é R$ 10,00.<br>
Se estiver chovendo, acrescenta R$ 2,00 à taxa padrão.<br><br>
O desafio desta atividade é criar um algoritmo em linguagem natural que informe o valor final da entrega.

```python
def delivery():
    km = int(input("Distancia do delivery até destino? "))
    clima = input("Verificar o clima, se está chovendo? (Sim/Não) ")
    taxa_chuva = 2
    taxa_5 = 5
    taxa_5_10 = 8
    taxa_10 = 10

    if km == 5 and clima == "Sim":
        taxa_chuva += taxa_5
        return print(f"Chovendo: {clima} - Taxa de entrega: {km}km: R${taxa_chuva}")    
    elif km == 5 and clima == "Não":
        return print(f"Chovendo: {clima} - Taxa de entrega: {km}km: R${taxa_5}")
        
    if km >= 5 and km <=10 and clima == "Sim":
        taxa_chuva += taxa_5_10
        return print(f"Chovendo: {clima} - Taxa de entrega: {km}km: R${taxa_chuva}")    
    elif km >= 5 and km <=10 and clima == "Não":
        return print(f"Chovendo: {clima} - Taxa de entrega: {km}km: R${taxa_5_10}")
    
    if km > 10 and clima == "Sim":
        taxa_chuva += taxa_10
        return print(f"Chovendo: {clima} - Taxa de entrega: {km}km: R${taxa_chuva}")    
    elif km == 5 and clima == "Não":
        return print(f"Chovendo: {clima} - Taxa de entrega: {km}km: R${taxa_10}")
    
        
```

Você está desenvolvendo um sistema de apoio para uma agência de viagens. Uma das funcionalidades mais solicitadas é um conversor de moedas. O usuário informa um valor em reais (R$) e o sistema precisa mostrar quanto isso representa em dólares (US$), usando uma taxa de câmbio definida pela empresa.

Sua tarefa é criar um algoritmo em linguagem natural que use uma função para fazer essa conversão. A função deve receber o valor em reais e a taxa de câmbio como entrada, e retornar o valor convertido.

```python
def conversorMoeda():

    valorReal = float(input("Digite o valor em reais (R$) que deve ser convertido: "))
    valorDolar = float(input("Taxa de câmbio (U$) hoje: "))
    valorConvert = (valorReal*valorDolar)

    return print(f"Valor de R${valorReal} convertido, com taxa de cambio U${valorDolar}. Ficará em U${valorConvert} .")

```

```python
conversorMoeda()
```

Você foi contratado por uma cafeteria que deseja automatizar o atendimento no balcão. O sistema deve permitir que o atendente registre os pedidos de cada cliente, calcule o valor total e aplique um desconto de 10% para clientes cadastrados.<BR>

O processo deve funcionar da seguinte forma:<BR>

O atendente informa quantos itens o cliente vai pedir.<BR>
Para cada item, o sistema solicita o nome e o preço.<BR>
Ao final, o sistema pergunta se o cliente é cadastrado.<BR>
Se for, aplica o desconto e exibe o valor com desconto.<BR>
Caso contrário, exibe o valor cheio.<BR>
O desafio consiste em criar um algoritmo que represente essa lógica de forma completa.

```python
cardapio = {'cafe': 4.00, 
            'pao':1.30,
            'refrigerante':5.00,
            'pingado': 6.00}

def cafeteria_total(lista):
    cadastro = input("Já possui Cadastro no sistema? (Sim/Não)")
    Desconto = 0.10
    total = 0
    itemNaoEnc = []
    
    for item in lista:
        if item in cardapio:
            total += cardapio[item]

    
    if total == 0:
        itemNaoEnc.append(item)
        print(f'Item não encontrado: {itemNaoEnc}')
    elif cadastro == "Sim":
        totalDesconto = total - (total*Desconto)
        print(f"Cliente com desconto, valor a pagar R${totalDesconto}")
    elif cadastro == "Não":
        print(f"Cliente sem desconto, valor a pagar R${total}")


cafeteria_total(['pao', 'pingado'])
```
